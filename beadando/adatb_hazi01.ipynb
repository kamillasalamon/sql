# **Házi feladat -** adatprofilozás a TERMÉK táblában

## **1\. lekérdezés**: Sorok száma

```sql
SELECT COUNT(*) AS 'sorok száma'
FROM Termek
```

## **2\. lekérdezés:** Hiányzó és nem hiányzó leírások

```sql
SELECT sum(iif(LEIRAS is null,1,0)) as 'Hiányzik a leírás',
       sum(iif(LEIRAS is not null,1,0)) as 'Rendelkezik leírással'
FROM Termek

```

## **3\. lekérdezés:** Átlagos listaár

```sql
SELECT AVG(LISTAAR) as 'Átlagos listaár'
FROM Termek
```

## **4\. lekérdezés:** Legdrágább és legolcsóbb termék

```sql
SELECT MIN(LISTAAR) as 'Legkisebb listaár',
       MAX(LISTAAR) as 'Legnagyobb listaár'
FROM Termek
```

## **5\. lekérdezés**: Ártartományok

```sql
SELECT SUM(CASE WHEN LISTAAR BETWEEN 0 AND 2000 THEN 1 ELSE 0 END) AS '0 és 2000 közötti listaár',
       SUM(CASE WHEN LISTAAR BETWEEN 2001 AND 4000 THEN 1 ELSE 0 END) AS '2001 és 4000 közötti listaár',
       SUM(CASE WHEN LISTAAR BETWEEN 4001 AND 6000 THEN 1 ELSE 0 END) AS '4001 és 6000 közötti listaár',
       SUM(CASE WHEN LISTAAR BETWEEN 6001 AND 8000 THEN 1 ELSE 0 END) AS '6001 és 8000 közötti listaár',
       SUM(CASE WHEN LISTAAR BETWEEN 8001 AND 10000 THEN 1 ELSE 0 END) AS '8001 és 10000 közötti listaár',
       SUM(CASE WHEN LISTAAR > 10000 THEN 1 ELSE 0 END) AS '10000 feletti listaár' 
FROM Termek
```

## **6\. lekérdezés**: Legkorábbi és legkésőbbi felvitel dátumai

```sql
SELECT MIN(FELVITEL) AS 'Legkorábban felvitt', 
       MAX(FELVITEL) AS 'Legkésőbb felvitt'
FROM Termek
```

## **7\. lekérdezés:** Felvitel hónapja szerinti eloszlás

```sql
SELECT month(FELVITEL) AS 'Felvitel hónapja',
       COUNT(*) AS 'db'
FROM Termek
GROUP BY month(FELVITEL)
ORDER BY month(FELVITEL)
```

## **8\. lekérdezés:** Terméket felvivők szerinti eloszlás

```sql
SELECT FELVITTE, 
       count(*) AS 'db'
FROM Termek
GROUP BY FELVITTE 
ORDER by count(*) asc
```
